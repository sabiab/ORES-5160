```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/CDC Diabetes Survillience burden percentage (2001-2020).csv")
data
```

```python
data.drop_duplicates()
```

```python
 df = data.drop(columns = "Country")
 df
```

```python
df1 = df.drop(columns = "State Code")
df1
```

```python
df1.info()
```

```python
df1['State'] = df1['State'].astype('category')
df1['County'] = df1['County'].astype('category')
df1.info()
```

```python
df1.isna().sum()
```

```python
df2 = df1.dropna()
df2
```

```python
df3 = df2.rename(columns = {
    "CountyFIPS" : "FIPS CODE",
    "Lower Limit" : "Lower Confidence Limit",
    "Upper Limit" : "Upper Confidence Limit"
 })
df3
```

```python
df3['Confidence Limit'] = df3['Upper Confidence Limit'] - df3['Lower Confidence Limit']
df3
```

```python
df3.drop_duplicates()
```

```python
sns.boxplot(df3['Percentage'])
```

```python
Q1 = df3['Percentage'].quantile(0.25)
Q3 = df3['Percentage'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df3[(df3['Percentage'] < lower_bound) | (df3['Percentage'] > upper_bound)]
outliers
```

```python
df4 = df3[~((df3['Percentage'] < lower_bound) | (df3['Percentage'] > upper_bound))]
df4
```

```python
df4.to_csv("CDC_Updated.csv", index=False)
```
